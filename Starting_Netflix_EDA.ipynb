For the rating column, we fill in missing values based on the type of the show. We assign the mode of rating for movies and TV shows separately.

# Finding the mode rating for movies and TV shows

````python

movie_rating = Netflix_df.loc[Netflix_df['type'] == 'Movie', 'rating'].mode()[0]
tv_rating = Netflix_df.loc[Netflix_df['type'] == 'TV Show', 'rating'].mode()[0]
````

# Filling missing rating values based on the type of content

````python

Netflix_df['rating'] = Netflix_df.apply(lambda x: movie_rating if x['type'] == 'Movie' and pd.isna(x['rating'])
                        else tv_rating if x['type'] == 'TV Show' and pd.isna(x['rating'])
                        else x['rating'], axis=1)
````

````python
Netflix_df["rating"][[5541,5794,5813]] #corrected the null values in rating
````

# Filling missing values in the 'country' column with the mode value

````python
Netflix_df['country'] = Netflix_df['country'].fillna(Netflix_df['country'].mode()[0])
````

# Replacing missing values in the 'director' column with 'No Data'

````python
Netflix_df['director'].fillna(Netflix_df['director'].mode()[0], inplace=True)
````

# Replacing missing values in the 'cast' column with 'No Data'

````python
Netflix_df['cast'].fillna(Netflix_df['cast'].mode()[0], inplace=True)
````

# Finding the mode duration for movies and TV shows

````python
movie_duration_mode = Netflix_df.loc[Netflix_df['type'] == 'Movie', 'duration'].mode()[0]
tv_duration_mode = Netflix_df.loc[Netflix_df['type'] == 'TV Show', 'duration'].mode()[0]
````

# Filling missing duration values based on the type of content

````python
Netflix_df['duration'] = Netflix_df.apply(lambda x: movie_duration_mode if x['type'] == 'Movie'
                          and pd.isna(x['duration'])
                          else tv_duration_mode if x['type'] == 'TV Show'
                          and pd.isna(x['duration'])
                          else x['duration'], axis=1)
````

````python
# Calculate the percentage distribution of content types
x = Netflix_df.groupby(['type'])['type'].count()
y = len(Netflix_df)
r = ((x/y) * 100).round(2)

ratio=pd.DataFrame(r)
ratio.columns=['Percentage']
ratio.index.name='Type'

#plotting pie chart
plt.figure(figsize=(5, 8))
colors = ['#b20710', '#221f1f']
explode = (0.1, 0)
plt.pie(ratio['Percentage'], labels=ratio.index, autopct='%1.1f%%',
colors=colors, explode=explode, shadow=True, startangle=90,
textprops={'color': 'white'})

plt.legend(loc='upper right')
plt.title('Distribution of Content Types')
plt.show()
plt.savefig("Distribution of Content Types.png")
````

# Best Week to launch a TV show/Movie?

````python
df1.head(3)
````

````python
df1['week_added'] = df1['date_added'].dt.isocalendar().week
````

````python
week = df1.groupby('week_added')['type'].value_counts()
week.name = 'count' # to avoid error while doing reset_index
week = week.reset_index()
week.info()
````

````python
#changing the week_added datatype to int

week['week_added'] = week['week_added'].astype('int64')

week.info()
````

````python
# creating two different tables for movies and tv shows
week_movie = week.loc[week['type'] == 'Movie'].sort_values(by = 'week_added')
week_tv = week.loc[week['type'] == 'TV Show'].sort_values(by = 'week_added')

#setting the plot style
fig = plt.figure(figsize = (15,7))
gs = fig.add_gridspec(2,1)

# creating graph for count of movies
ax0 = fig.add_subplot(gs[0,:])
ax0.set_facecolor('#f6f5f5')
sns.lineplot(data = week_movie, x = 'week_added', y = 'count',marker = 'o',markersize = 10,color ='#b20710',
             label = 'Movie',ax = ax0)

#customizing axis labels
ax0.set_xlabel('Week Added',fontweight = 'bold',fontfamily='serif')
ax0.set_ylabel('Count',fontweight = 'bold',fontfamily='serif')


# creating graph for count of tv shows
ax1 = fig.add_subplot(gs[1,:])
ax1.set_facecolor('#f6f5f5')
sns.lineplot(data = week_tv, x = 'week_added', y = 'count',marker = 'o',markersize = 10,color ='#221f1f',
             label = 'TV Show', ax = ax1)

#customizing axis labels
ax1.set_xlabel('Week Added',fontweight = 'bold',fontfamily='serif')
ax1.set_ylabel('Count',fontweight = 'bold',fontfamily='serif')

#creating the title
ax0.set_title('Best Week to launch a Movie/TV show?',
            {'font':'serif', 'size':15,'weight':'bold'})


plt.show()
plt.savefig("Best Week to launch a Movie or TV show")
````

````python
#function for binning age groups
def age_group(x):
    if x in ['TV-Y', 'TV-G', 'G']:
        return 'Little Kids'
    elif x in ['TV-Y7', 'TV-Y7-FV', 'TV-PG', 'PG']:
        return 'Older Kids'
    elif x in ['PG-13','TV-14']:
        return 'Teens'
    else:
        return 'Adults'

#creating a df for binning the age groups
#binning ratings into age groups for movies
movie_rating['age'] = movie_rating['rating'].apply(lambda x:age_group(x))
````

````python
age_m_rating = movie_rating.groupby('age')['count'].sum().sort_values(ascending = False).reset_index()
````

#binning ratings into age groups for tv shows

````python
tv_rating['age'] = tv_rating['rating'].apply(lambda x:age_group(x))
age_t_rating = tv_rating.groupby('age')['count'].sum().sort_values(ascending = False).reset_index()

age_m_rating
````

````python
df1['duration'] = df1['duration'].str.extract('(\d+)', expand=False).astype(int)
````

````python
# Extracting and converting the duration for movies
df_movies=df1[df1['type']=='Movie']
df_tv_shows=df1[df1['type']=='TV Show']
````

````python
# Creating a boxplot for movie duration
fig, axes = plt.subplots(1, 2, figsize=(14, 6)) # 1 row, 2 columns
sns.boxplot(data=df_movies, x='type', y='duration', color='#b20710', ax=axes[0])
axes[0].set_xlabel('Content Type')
axes[0].set_ylabel('Duration')
axes[0].set_title('Distribution of Duration for Movies')

# Creating a boxplot for TV show duration
sns.boxplot(data=df_tv_shows, x='type', y='duration', color='#221f1f', ax=axes[1])
axes[1].set_xlabel('Content Type')
axes[1].set_ylabel('Duration')
axes[1].set_title('Distribution of Duration for TV Shows')
````

*What we can infer from above :*

* Analyzing the movie box plot, we can see that most movies fall within a reasonable duration range, with few outliers exceeding approximately 2.5 hours. This suggests that most movies on Netflix are designed to fit within a standard viewing time.

* For TV shows, the box plot reveals that most shows have one to four seasons, with very few outliers having longer durations. This aligns with the earlier trends, indicating that Netflix focuses on shorter series formats.
